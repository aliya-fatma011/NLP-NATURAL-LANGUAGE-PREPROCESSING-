#REMOVE PUNCTUATION

Removing Punctuation in NLP

Removing punctuation is an important step in text preprocessing because punctuation marks usually do not add meaning for most NLP tasks.

✅ Why Remove Punctuation?

Punctuation like:

! , . ? : ; " ' ( ) [ ] { }


can create unnecessary tokens and increase vocabulary size.

Example:

“hello!”

“hello”

These are treated as two different words unless punctuation is removed.

```python
import string
exclude = string.punctuation
def remove_punc(text):
  for char in exclude:
    text=text.replace(char,'')
  return text
```

#Fast spelling checker

```python

from spellchecker import SpellChecker

spell = SpellChecker()

def spelling_correction(text):
    corrected_words = []
    for word in text.split():
        corrected_words.append(spell.correction(word) or word)
    return " ".join(corrected_words)
```

```python
text = "this moovie was amazng but story waz boring"
print(spelling_correction(text))

```

#remove emojis from text

```python

import re

def remove_emoji(text):
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags
        u"\U00002700-\U000027BF"
        u"\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)
```

```python

text = "This movie was amazing 😍🔥 but ending was bad 😢"
print(remove_emoji(text))
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
def remove_stopwords(text):
  new_text=[]
  for word in text.split():
    if word in stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(word)
  x=new_text[:]
  new_text.clear()
  return " ".join(x)
```

```python
remove_stopwords('i am not a very good cricket player')
```

STEMMING – Explanation (Simple & Clear)

Stemming is a technique in Natural Language Processing (NLP) used to reduce words to their base or root form.

It removes prefixes or suffixes from a word so that different forms of a word are treated as the same.

🔹 Definition
Stemming is the process of converting a word to its root form by chopping off its affixes (suffixes or prefixes).

🔹 Why Stemming is Used?
In text processing, the same word can appear in many forms.

Example:

play

playing

played

plays

All these words have the same meaning.
Stemming converts all of them into one root word:

👉 play

This helps computers understand that they are the same word.

🔹 Examples of Stemming
Original Word	Stemmed Word
running	run
studies	studi
connected	connect
easily	easili
You can see that stemming does not always produce a real dictionary word (like “studi”, “easili”).

🔹 How Stemming Works?
Stemming uses simple rule-based algorithms.

Most popular algorithm:

👉 Porter Stemmer

It removes common endings like:

-ing

-ed

-s

-es

-ly

🔹 Difference Between Stemming and Lemmatization
Stemming	Lemmatization
Fast and simple	Slower but accurate
Rule-based	Dictionary-based
May produce incorrect words	Produces meaningful words
Example:

Better → Stemming: better

Better → Lemmatization: good

🔹 Use of Stemming
Stemming is mainly used in:

Search engines

Text classification

Sentiment analysis

Spam filtering

Chatbots

🟢 Simple Summary
Stemming = Cutting words to their root form to simplify text processing.

Lemmatization is a text preprocessing technique used in Natural Language Processing (NLP).

Simple Definition
Lemmatization means converting a word to its base or dictionary form, called a lemma.

Easy Example
Word	Lemma (Base Form)
running	run
better	good
went	go
children	child
So, lemmatization changes different forms of a word into one common base form.

Why is Lemmatization Used?
In NLP, the same word can appear in many forms.
For example:

play

playing

played

plays

All these have the same meaning: “play”

Lemmatization helps the computer understand that all these words are actually the same.

How Lemmatization Works
Lemmatization is intelligent.
It uses:

Vocabulary

Grammar rules

Parts of speech (POS)

Because of this, it gives meaningful base words, not just cutting endings.

Lemmatization vs Stemming
Stemming	Lemmatization
Just removes word endings	Uses dictionary and grammar
Faster but less accurate	Slower but more accurate
Example: “studies” → “studi”	“studies” → “study”
Lemmatization gives real words, while stemming may give incorrect forms.

Example in Python (using NLTK)
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("running", pos="v"))
Output:

run
In Short
👉 Lemmatization = Converting words into their meaningful base form

It helps in:

Text analysis

Sentiment analysis

Chatbots


```python
import nltk
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
sentence='he was eating and running at the same time.he has bad habitof swimming after playing long hours in the sun'
punctuations="?:.,!;"
sentence_words=nltk.word_tokenize(sentence)
for word in sentence_words:
  if word in punctuations:
    sentence_words.remove(word)
sentence_words
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words:
  print("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word)))
```
